```python
import pymysql

class PythonMYSQL():
    def __init__(self):
        self.db = pymysql.connect(host='localhost',
                                 user='pydb',
                                 password='mariadb',
                                 db='pydb',
                                 charset='utf8')
        
        self.cursor = self.db.cursor(pymysql.cursors.DictCursor)
        
    def execute(self, query, args={}):
        self.cursor.execute(query, args) 
 
    def executeOne(self, query, args={}):
        self.cursor.execute(query, args)
        row= self.cursor.fetchone()
        return row
 
    def executeAll(self, query, args={}):
        self.cursor.execute(query, args)
        row= self.cursor.fetchall()
        return row
    
    def commit():
        self.db.commit()
```

```python
from flask import Blueprint, request, render_template, flash, redirect, url_for
from flask import current_app as current_app
from flask import Flask

app = Flask(__name__)

@app.route('/index')
@app.route('/', methods=['GET'])
def index():
    return render_template('index.html',
                            result=None,
                            resultData=None,
                            resultUPDATE=None)
 
 
 
# INSERT 함수 예제
@app.route('/insert', methods=['GET'])
def insert():
    db_class= PythonMYSQL()
 
    sql     = "INSERT INTO pydb.pythonTable(test) \
                VALUES('%s')"% ('testData')
    db_class.execute(sql)
    db_class.commit()
 
    return render_template('/index.html',
                           result='insert is done!',
                           resultData=None,
                           resultUPDATE=None)
 
 
 
# SELECT 함수 예제
@app.route('/select', methods=['GET'])
def select():
    db_class= PythonMYSQL()
 
    sql     = "SELECT idx, test \
                FROM pydb.pythonTable"
    row     = db_class.executeAll(sql)
 
    print(row)
 
    return render_template('/index.html',
                            result=None,
                            resultData=row[0],
                            resultUPDATE=None)

# UPDATE 함수 예제
@app.route('/update', methods=['GET'])
def update():
    db_class= PythonMYSQL()
 
    sql     = "UPDATE pydb.pythonTable \
                SET test='%s' \
                WHERE test='testData'"% ('update_Data')
    db_class.execute(sql)   
    db_class.commit()
 
    sql     = "SELECT idx, test \
                FROM pydb.pythonTable"
    row     = db_class.executeAll(sql)
 
    return render_template('/index.html',
                            result=None,
                            resultData=None,
                            resultUPDATE=row[0])

app.run()
```
